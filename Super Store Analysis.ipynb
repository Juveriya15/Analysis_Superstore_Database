```python
#Getting month year fron the dataset
df['month_year']=df['order_date'].apply(lambda x:x.strftime('%Y-%n'))
```

```python
#Grouping month year
df_trend=df.groupby('month_year').sum()['sales'].reset_index()
```

```python
#Visualize
plt.figure(figsize=(15,6))
plt.plot(df_trend['month_year'],df_trend['sales'], color='red')
```

```python
#WHICH ARE THE TOP 10 PRODUCTS?
```

```python
#Grouping product name column
product_sales=pd.DataFrame(df.groupby('product_name').sum()['sales'])
```

```python
#Sorting Product Sales
product_sales.sort_values('sales',ascending=False)
```

```python
#TOP 10 Products by sales
product_sales[:10]
```

```python
#WHICH ARE THE MOST SELLING PRODUCTS?

# Grouping product name
most_selling_product= pd.DataFrame(df.groupby('product_name').sum()['quantity'])
```

```python
#Sorting Most Selling products
most_selling_product= most_selling_product.sort_values('quantity', ascending=False)
```

```python
most_selling_product[:10]
```

```python
#WHAT IS THE MOST PREFERED SHIP MODE
#setiing figure size
plt.figure(figsize=(10, 8))

# countplot: Show the counts of observations in each categorical bin using bars
sns.countplot(x='ship_mode', data=df)

#display figure
plt.show()
```

```python
#WHICH ARE THE MOST PROFITABLE CATEGORY?

#Grouping Category and Subcategory
catory_sub_profit= pd.DataFrame(df.groupby(['category','sub_category']).sum()['profit'])
```

```python
#Sorting the results
catory_sub_profit.sort_values(['category','profit'],ascending=False)
```
